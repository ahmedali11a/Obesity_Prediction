# Handling Outliers in Numerical Features

Detect and treat outliers using the Interquartile Range (IQR) method by capping values outside the acceptable range. This helps reduce the impact of extreme values on model performance.

```python
def Lower_Upper(col):
    q1, q3 = np.percentile(col, [25, 75])
    iqr = q3 - q1
    Lower = q1 - 1.5 * iqr
    Upper = q3 + 1.5 * iqr
    return Lower, Upper
```

```python
for col in numeric_train_cols:
    lw,up=Lower_Upper(df_train[col])
    df_train[col]=np.where(df_train[col]<lw,lw,df_train[col])
    df_train[col]=np.where(df_train[col]>up,up,df_train[col])
```

```python
for col in numeric_train_cols:
    lw, up = Lower_Upper(df_train[col]) 
    df_test[col] = np.where(df_test[col] < lw, lw, df_test[col])
    df_test[col] = np.where(df_test[col] > up, up, df_test[col])
```

# Model Training, Evaluation, and Saving

Train multiple classification models on the training data, evaluate their performance on both training and test sets using accuracy, precision, recall, F1-score, confusion matrix, and cross-validation accuracy.

```python
model_dir = 'models'
os.makedirs(model_dir, exist_ok=True)




target_col = 'NObeyesdad'
y_train = df_train[target_col]
X_train = df_train.drop(columns=[target_col])

y_test = df_test[target_col]
X_test = df_test.drop(columns=[target_col])

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=20,           
    min_samples_split=5,    
    min_samples_leaf=2,     
    max_features='sqrt',    
    random_state=42
)

gb = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.05,      
    max_depth=5,            
    subsample=0.8,               
    random_state=42
)


models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM": SVC(kernel='rbf', C=1, gamma='scale', probability=True),
    "Random Forest": rf,
    "Gradient Boosting": gb
}

results = []

for name, model in models.items():
    print(f"--- {name} ---")
    
    model.fit(X_train, y_train)

    model_filename = os.path.join(model_dir, f"{name.lower().replace(' ', '_')}.pkl")
    joblib.dump(model, model_filename)
    print(f"Saved model to {model_filename}")
    
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)
    prec = precision_score(y_test, y_test_pred, average='macro')
    rec = recall_score(y_test, y_test_pred, average='macro')
    f1 = f1_score(y_test, y_test_pred, average='macro')
    cm = confusion_matrix(y_test, y_test_pred)
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    
    print(f"Train Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")
    if train_acc - test_acc < 0.15:
        print("There is not overfitting :)")
    else:
        print("Unfortunately, there is overfitting :(")
    print(f"Precision (macro avg): {prec:.4f}")
    print(f"Recall (macro avg): {rec:.4f}")
    print(f"F1-score (macro avg): {f1:.4f}")
    print(f"Confusion Matrix:\n{cm}")
    print(f"Cross-Validation Accuracy (5-Fold): {cv_scores.mean():.4f}")
    print("="*50)
    
    results.append({
        'Model': name,
        'Train Accuracy': train_acc,
        'Test Accuracy': test_acc,
        'Precision': prec,
        'Recall': rec,
        'F1-score': f1,
        'CV Accuracy Mean': cv_scores.mean()
    })

results_df = pd.DataFrame(results)
print("\n=== Summary of All Models ===")
print(results_df)
```

# Hyperparameter Tuning for XGBoost Classifier

Perform manual grid search over different combinations of `max_depth` and `learning_rate` to find the best model configuration based on test accuracy. 

```python
best_accuracy = 0
best_train_accuracy = 0
best_params = None
best_model = None

param_trials = [
    {'max_depth': 3, 'learning_rate': 0.1},
    {'max_depth': 6, 'learning_rate': 0.1},
    {'max_depth': 3, 'learning_rate': 0.01},
    {'max_depth': 6, 'learning_rate': 0.01},
]

for trial_num, params in enumerate(param_trials, start=1):
    model = XGBClassifier(
        eval_metric='mlogloss',
        max_depth=params['max_depth'],
        learning_rate=params['learning_rate'],
        random_state=42
    )
    
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    train_acc = accuracy_score(y_train, y_train_pred)
    
    y_test_pred = model.predict(X_test)
    test_acc = accuracy_score(y_test, y_test_pred)
    
    print(f"Trial {trial_num}: max_depth={params['max_depth']}, learning_rate={params['learning_rate']}")
    print(f"Train Accuracy: {train_acc:.4f} | Test Accuracy: {test_acc:.4f}")
    print("="*50)
    
    if test_acc > best_accuracy or (test_acc == best_accuracy and train_acc > best_train_accuracy):
        best_accuracy = test_acc
        best_train_accuracy = train_acc
        best_params = params
        best_model = model

print("\nBest Model Summary:")
print(f"Best Parameters: max_depth={best_params['max_depth']}, learning_rate={best_params['learning_rate']}")
print(f"Best Test Accuracy: {best_accuracy:.4f}")
print(f"Best Train Accuracy: {best_train_accuracy:.4f}")
```

# Obesity Level Prediction

```python
model = joblib.load('models/gradient_boosting.pkl')
preprocessing = joblib.load('models/preprocessing.pkl')

label_encoders = preprocessing['label_encoders']
scalers = preprocessing['scaler']
final_columns = preprocessing['final_columns']

feature_columns = final_columns.copy()
if 'NObeyesdad' in feature_columns:
    feature_columns.remove('NObeyesdad')
if 'BMI' in feature_columns:
    feature_columns.remove('BMI')  

user_input = {}
print("Please enter the following data:")

def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

for col in feature_columns:
    is_categorical = col in label_encoders
    while True:
        val = input(f"{col}: ").strip()
        
        if is_categorical:
            if is_number(val):
                print(" Error: Add Characters only")
                continue
            val = val.lower()
            if val in ['yes', 'no']:  
                pass
            else:
                val = val.replace("_", " ").title().replace(" ", "_")
            user_input[col] = val
            break
        else:
            if not is_number(val):
                print(" Error: Add Numbers only")
                continue
            user_input[col] = float(val)
            break

user_df = pd.DataFrame([user_input])

if 'Weight' in user_df.columns and 'Height' in user_df.columns:
    user_df['BMI'] = user_df['Weight'] / (user_df['Height'] ** 2)

for col, le in label_encoders.items():
    if col in user_df.columns:
        user_df[col] = user_df[col].apply(lambda x: x if x in le.classes_ else le.classes_[0])
        user_df[col] = le.transform(user_df[col])

for col in scalers.keys():
    if col in user_df.columns:
        user_df[col] = scalers[col].transform(user_df[[col]])

user_df = user_df[final_columns]

prediction = model.predict(user_df)

print("\n Predicted Obesity Category:")
classes = [
    "Insufficient_Weight", "Normal_Weight", "Obesity_Type_I",
    "Obesity_Type_II", "Obesity_Type_III", "Overweight_Level_I",
    "Overweight_Level_II"
]
if prediction[0] in range(len(classes)):
    print(classes[prediction[0]])
else:
    print("Unknown category")
```
